```python
def blablo(csv_path):
    df1 = pd.read_csv(csv_path)
    regression = LinearRegression()
    for z in range(15,32):
        X_train = df1.iloc[:15,1:2].values # all cols and rows except the last(quality)
        Y_train = df1.iloc[:15,2:3].values #the col quality
        X_test = df1.iloc[z:(z+1),1:2].values # all except type


        # Train the model
        regression.fit(X_train, Y_train)

        # Predict using the trained model

        predictions = regression.predict(X_test)
        a = predictions
        #print(type(a))
        b = round(a[0][0],2)
        # Define the data to be written
        data = b

        # Open the CSV file in read mode
        with open(csv_file, 'r', newline='') as file:
            reader = csv.reader(file)
            rows = list(reader)

        # Update the value at row 15, column 3
        rows[z+1][2] = data
        rows[z+2][1] = data

        # Open the CSV file in write mode
        with open(csv_file, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(rows)
            #writer.writerows(rowss)
        print(data)
            
blablo("C:\\Users\\Admin\\testttt_A.csv")
    
    
```
